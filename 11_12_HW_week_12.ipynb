## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
# Create a simple plot of the values
plt.scatter(x, y)   # looks linear

# z = new calculation of a linear trendline for our data
z = np.polyfit(x, y, 1)

# p = plot of that line
p = np.poly1d(z)

# combine both plots
plt.plot(x, p(x))
plt.scatter(x,y)
```
